```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = 'C:\\Users\\ASHWITHA RAMANOLLA\\Desktop\\DataAnalysis\\amazon.csv'
data = pd.read_csv(file_path)

# Basic Analysis
print(data.head())
print(data.describe())
print(data.isnull().sum())

# Verify column names
print(data.columns)

# Visualization - Histogram of 'rating'
plt.figure(figsize=(10, 6))
sns.histplot(data['rating'])
plt.title('Distribution of Rating')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.savefig('C:\\Users\\ASHWITHA RAMANOLLA\\Desktop\\DataAnalysis\\histogram_rating.png')
plt.show()

# Visualization - Scatter Plot of 'rating' vs 'rating_count'
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['rating'], y=data['rating_count'])
plt.title('Scatter plot of Rating vs Rating Count')
plt.xlabel('Rating')
plt.ylabel('Rating Count')
plt.savefig('C:\\Users\\ASHWITHA RAMANOLLA\\Desktop\\DataAnalysis\\scatter_plot_rating_vs_rating_count.png')
plt.show()

# Save summary statistics
summary = data.describe()
summary.to_csv('C:\\Users\\ASHWITHA RAMANOLLA\\Desktop\\DataAnalysis\\summary_statistics.csv')

# Save cleaned data (if any cleaning is done)
data.to_csv('C:\\Users\\ASHWITHA RAMANOLLA\\Desktop\\DataAnalysis\\cleaned_data.csv', index=False)
```
